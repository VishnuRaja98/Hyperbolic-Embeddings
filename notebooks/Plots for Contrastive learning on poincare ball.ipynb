```python
import json
import matplotlib.pyplot as plt
import numpy as np
import glob

def plot_spearman_by_dimension(output_file='spearman_comparison.png'):
    """
    Plots validation Spearman scores across epochs for different embedding dimensions.
    
    Args:
        log_dir: Directory containing training_logs_{label}.json files
        output_file: Filename to save the plot
    """
    plt.figure(figsize=(12, 7))
    hyp_dims = [384,192,96,64,48,32,24,16,12,8]
    for dim in hyp_dims:
        log_file = f"./model_logs2/hyp_{dim}_logs.json"
        
        try:
            with open(log_file, 'r') as f:
                data = json.load(f)
            
            # Extract Spearman scores and epochs
            epochs = [e['epoch'] for e in data['epochs']]
            spearman = [-e['val_spearman'] for e in data['epochs']]
            
            # Plot curve
            plt.plot(epochs, spearman, 
                    marker='o', 
                    markersize=2,
                    linewidth=1,
                    label=f"Dim {dim}")
            
        except FileNotFoundError:
            print(f"Warning: Log file not found for dim {dim} - {log_file}")
            continue
    
    # Format plot
    plt.title('Hyperbolic Embedding Dimension Comparison\nValidation Spearman Score by Epoch', pad=20)
    plt.xlabel('Epoch', fontsize=1)
    plt.ylabel('Spearman Correlation', fontsize=12)
    plt.ylim(0.5, 0.8)  # Adjusted for typical Spearman ranges
    plt.xticks(np.arange(1, max(epochs)+1))
    
    # Customize appearance
    plt.grid(True, alpha=0.2, linestyle='--')
    plt.legend(title='Embedding Dimension', 
              bbox_to_anchor=(1.05, 1), 
              loc='upper left')
    plt.tight_layout()
    
    # Save and show
    plt.savefig(output_file, dpi=300, bbox_inches='tight')
    plt.show()
    print(f"Plot saved to {output_file}")

plot_spearman_by_dimension(output_file='spearman_comparison.png')
```

```python
def plot_train_loss_by_dimension(output_file='train_loss_comparison.png'):
    """
    Plots validation train loss across epochs for different embedding dimensions.
    
    Args:
        log_dir: Directory containing training_logs_{label}.json files
        output_file: Filename to save the plot
    """
    plt.figure(figsize=(12, 7))
    hyp_dims = [384,192,96,64,48,32,24,16,12,8]
    for dim in hyp_dims:
        log_file = f"./model_logs2/hyp_{dim}_logs.json"
        
        try:
            with open(log_file, 'r') as f:
                data = json.load(f)
            
            # Extract Spearman scores and epochs
            epochs = [e['epoch'] for e in data['epochs']]
            spearman = [e['train_loss'] for e in data['epochs']]
            
            # Plot curve
            plt.plot(epochs, spearman, 
                    marker='o', 
                    markersize=2,
                    linewidth=1,
                    label=f"Dim {dim}")
            
        except FileNotFoundError:
            print(f"Warning: Log file not found for dim {dim} - {log_file}")
            continue
    
    # Format plot
    plt.title('Hyperbolic Embedding Dimension Comparison\nTrain Loss by Epoch', pad=20)
    plt.xlabel('Epoch', fontsize=1)
    plt.ylabel('Train Loss', fontsize=12)
    plt.ylim(2,9)  # Adjusted for typical Spearman ranges
    plt.xticks(np.arange(1, max(epochs)+1))
    
    # Customize appearance
    plt.grid(True, alpha=0.2, linestyle='--')
    plt.legend(title='Embedding Dimension', 
              bbox_to_anchor=(1.05, 1), 
              loc='upper left')
    plt.tight_layout()
    
    # Save and show
    plt.savefig(output_file, dpi=300, bbox_inches='tight')
    plt.show()
    print(f"Plot saved to {output_file}")

plot_train_loss_by_dimension(output_file='train_loss_comparison.png')
```
